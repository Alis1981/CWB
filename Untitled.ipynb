```python
def compare_graphs(var_m, var_p):
    
    minimo = min(np.min(var_m), np.min(var_p))
    maximo = max(np.max(var_m), np.max(var_p))
    #first graph
    fig, (ax1, ax2, ax3) = plt.subplots(1,3, sharey=False)#, gridspec_kw={'height_ratios': [2, 2, 2]})
    divider = make_axes_locatable(ax1)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    ax1.set_title('Matlab')
    im1 = ax1.imshow(var_m, vmin = minimo, vmax = maximo)
    fig.colorbar(im1, cax=cax, orientation='vertical')
    #ax2.set_ylim(var_m.shape[0], var_m.shape[1])   
    #second graph
    ax2.set_title('Python')
    divider = make_axes_locatable(ax2)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    im2 = ax2.imshow(var_p, vmin = minimo, vmax = maximo)
    fig.colorbar(im2, cax=cax, orientation='vertical')

    #ax2.set_ylim(var_p.shape[0], var_p.shape[1])
    #plt.show()
    
    #third graph
    #ax3 = plt.subplots(1,3)
    #ax3 = fig.add_subplot(1, 3, 3)
    python = var_p.reshape(-1)
    matlab = var_m.reshape(-1)
    
    ax3.set_title('Comparison')
    ax3.scatter(python, matlab)
    ax3.plot([minimo,maximo],[minimo,maximo], 'r')
    ax3.set_xlabel('Python')
    ax3.set_ylabel('Matlab')
    ax3.set(adjustable='box', aspect='equal')
    ax3.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.1f}"))
    ax3.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.1f}"))
    plt.show()
```
