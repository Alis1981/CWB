```python
foto = imagen5[0:16,0:16]
dim = foto.shape
y, x = np.arange(0, dim[0]), np.arange(0, dim[1])
Y, X = np.meshgrid(y, x, indexing='ij')

u = np.cos(foto[Y,X])
v = np.sin(foto[Y,X])

plt.quiver(X,Y,u,v)
plt.axis('equal')
plt.show()
#Maximun Theta angle first dominant mode
```
